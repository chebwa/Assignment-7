```python
IBM_Attrition["AttritionCount"]= IBM_Attrition["Attrition"].map({"Yes": 1, "No": 0}) 
IBM_Attrition.head()
```

```python
Attrition_only= IBM_Attrition[IBM_Attrition["Attrition"]== "Yes"]
Attrition_only
```

```python
#Analysis1- Distance from home and Job Role

import matplotlib.pyplot as plt
fig,ax=plt.subplots(1,2)
ax[0].hist(Attrition_only['DistanceFromHome'], rwidth=0.8)
ax[0].set_title('Distance Frm Home')
ax[1].hist(Attrition_only['JobRole'], rwidth=0.8)
ax[1].set_title('Job Role')
plt.xticks(rotation = 'vertical')
plt.ylabel('Attrition count')
plt.show()
```

```python
#Provides Attrition rate by Job Role as a Percentage

JRbyAtt = Attrition_only.groupby(['JobRole']).agg({'EmployeeCount': 'sum'})
AttFile = IBM_Attrition.groupby(['JobRole']).agg({'EmployeeCount': 'sum'})
JRbyAtt.div(AttFile, level = "JobRole") * 100
```

```python
IBM_Attrition.groupby(["Attrition", "JobRole"])["DistanceFromHome"].mean()
```

```python
# Relationship between Distance from Home and Job Role
import seaborn as sns
g = sns.catplot(x = "JobRole", y= "DistanceFromHome", col= "Attrition", data = IBM_Attrition, kind ="bar")
g.set_xticklabels(rotation = 'vertical')
```

```python
JRTotal = IBM_Attrition["JobRole"]
AttbyJR = Attrition_only["JobRole"]
plt.hist([AttbyJR, JRTotal], label=["Att count by JobRole", "JR Total count"])
plt.legend()
plt.xticks(rotation = 'vertical')
plt.xlabel('Job Role')
plt.ylabel('Count')

plt.title(' Bar chart Graph')

plt.show()
```

```python
#Analysis2- Monthly Income and Education
fig,ax=plt.subplots(1,2)
ax[0].hist(Attrition_only['MonthlyIncome'], rwidth=0.8, label = 'Monthly Income')
ax[0].set_title('Monthly Income')
ax[1].hist(Attrition_only['Education'], label = 'Education')
ax[1].set_title('Education')
plt.ylabel('Attrition count')
plt.show()
```

```python
IBM_Attrition.groupby(["Attrition", "Education"])["MonthlyIncome"].mean()
```

```python
import seaborn as sns
sns.catplot(x="Education", y = "MonthlyIncome", hue= "Attrition", data= IBM_Attrition, kind= "point")
```

```python
#Analysis3- Performance Rating and Salary Hike

fig,ax=plt.subplots(1,2)
ax[0].hist(Attrition_only['PerformanceRating'], rwidth=0.8)
ax[0].set_title('Performance Rating')
ax[1].hist(Attrition_only['PercentSalaryHike'], rwidth=0.8)
ax[1].set_title('% Salary Hike')
plt.ylabel('Attrition count')
plt.show()
```

```python
IBM_Attrition.groupby(["PerformanceRating", "PercentSalaryHike"])["AttritionCount"].sum() 
```

```python
sns.catplot(x="PerformanceRating", y = "PercentSalaryHike", hue = "Attrition", data = IBM_Attrition, kind = "box")
```

```python
#Analysis4- Years at Company, Years since last Promotion and Job Satisfaction
fig,ax=plt.subplots(1,3)
ax[0].hist(Attrition_only['YearsAtCompany'], rwidth=0.8)
ax[0].set_title('Yrs at Comp')
ax[1].hist(Attrition_only['YearsSinceLastPromotion'], rwidth=0.8)
ax[1].set_title('Yrs since last Prom')
ax[2].hist(Attrition_only['JobSatisfaction'], rwidth=0.8)
ax[2].set_title('Job Satisafaction')
plt.ylabel('Attrition count')
plt.show()
```

```python
d= sns.catplot(x = "Department", y= "JobSatisfaction", col= "Attrition", data = IBM_Attrition, kind = "point")
d.set_xticklabels(rotation = 30)
```

```python
#Analysis5- Gender and Marital Status

fig,ax=plt.subplots(1,2)
ax[0].hist(Attrition_only['MaritalStatus'], rwidth=0.8)
ax[0].set_title('Marital Status')
ax[1].hist(Attrition_only['Gender'], rwidth=0.8)
ax[1].set_title('Gender')
plt.ylabel('Attrition count')
plt.show()
```

```python
#Attrition rate by Gender as a %

GenbyAtt = Attrition_only.groupby(['Gender']).agg({'EmployeeCount': 'sum'})
AttFile = IBM_Attrition.groupby(['Gender']).agg({'EmployeeCount': 'sum'})
GenbyAtt.div(AttFile, level = "Gender") * 100
```

```python
#Attrition rate by Marital Status as a %

GenbyAtt = Attrition_only.groupby(['MaritalStatus']).agg({'EmployeeCount': 'sum'})
AttFile = IBM_Attrition.groupby(['MaritalStatus']).agg({'EmployeeCount': 'sum'})
GenbyAtt.div(AttFile, level = "MaritalStatus") * 100
```
